## Check frequency spectra of our signals


This notebook takes the data from [Demo 1: Data Capture in caprini](demo_1_capture.ipynb) and looks for patterns in the spectra of the data.

We'll focus on channel 1, but could use whatever.

First, let's load up our datasets, and plot channel 1.

```python
with open("data_initial.json") as fh:
    wfb_initial = WaveformBundle.from_json(fh)
    
with open("data_ch1_connected.json") as fh:
    wfb_test_1 = WaveformBundle.from_json(fh)
```

```python
plot(wfb_initial.waveforms["CHAN1"].t, wfb_initial.waveforms["CHAN1"].y, label="initial")
plot(wfb_test_1.waveforms["CHAN1"].t, wfb_test_1.waveforms["CHAN1"].y, label="test1")
legend()

title("Voltages present on Channel 1 in our experiments")
ylabel("Volts")
xlabel("Time (s)");
```

That looks reasonable.  Now, let's look at the power spectra of these two.

```python
psd(wfb_initial.waveforms["CHAN1"].y, label="Initial") 
psd(wfb_test_1.waveforms["CHAN1"].y, label="Test 1")
legend();
```
